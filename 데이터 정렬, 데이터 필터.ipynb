```python
### 데이터 필터
```

```python
df1['연령대'].value_counts()
```

```python
df1['연령대'] == '40대'
```

```python
cond1 = (df1['연령대'] == '40대')
df1.loc[cond1] ## 연령대에서 40대의 데이터만 불러옴
```

```python
cond2 = (df1['연령대'] != '40대')
df1.loc[cond2] ## 40대를 제외한 데이터만 불러옴
```

```python
df1.loc[~cond1] ## 위와 같이 40대를 제외한 데이터들만 불러옴
```

```python
df1['성별'].value_counts()
```

```python
cond1 = (df1['연령대'] == '40대')
```

```python
cond2 = (df1['성별'] == '남')
```

```python
df1.loc[cond1 & cond2] ## 성별이 남성이면서 연령대가 40대의 사람만 불러옴
```

```python
df1.loc[cond1 | cond2]
```

```python
# 30대 여성 추출 > 구매금액 높은순서로 정렬
```

```python
# 하나의 파일로 저장
```

```python
cond1 = (df1['연령대'] == '30대이하')
cond2 = (df1['성별'] == '여')
```

```python
df3 = df1[cond1 & cond2].sort_values(by='구매금액', ascending=False)
```
