```python
df2 = pd.read_csv('Data02.csv')
print(df2.shape)
df2
```

```python
pd.to_datetime(df2['배송시작일']) ## 오전같은 한글 글자 때문에 오류 발생
```

```python
def date_function(x):
    return x[0:10]

df2['배송시작일'].apply(date_function)  ## 가장 앞에있는 숫자부터 10번째 전 까지의 데이터를 잘라온다
```

```python
df2['배송시작일(datetime)'] = pd.to_datetime(df2['배송시작일'].apply(date_function))
df2['배송완료일(datetime)'] = pd.to_datetime(df2['배송완료일'].apply(date_function))
```

```python
df2
```

```python
df2['배송시간'] = df2['배송완료일(datetime)'] - df2['배송시작일(datetime)'] ## 시간의 간격을 나타내는 데이터
```

```python
df2['주문일'] ## 데이터가 붙어있어 숫자로 인식
```

```python
def date_function2(x):
    result = str(x) ## 문자열로 바꾼 후
    return result[0:4] + '-' + result[4:6] + '-' + result[6:8] ## 연 월 일을 나눠준다

df2['주문일'].apply(date_function2) ## 이제 위와 같이 to_datetime 사용 가능
```
