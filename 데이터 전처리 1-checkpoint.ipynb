```python
df6 = df1[['카테고리명','자재그룹','자재그룹명','제품코드','제품명','안전재고',' 분류']]
df6
```

```python
# Table Join
# - right / left # 오른쪽, 왼쪽 기준으로 붙힌다 (없는 데이터는 NaN(결측치)으로 표시)
# - inner # 중복되어 있는 데이터만 가져와 붙힌다 A & B
# - outer # 중복되어 있지 않은 데이터들도 가져와 붙힌다( 없는 데이터는 NaN으로 표시 ) A or B
```

```python
df7 = pd.merge(df6,df5, how='right', on='제품명') ## df5를 df6에 붙힘
df7
```

```python
df7.to_excel('total_Data01.xlsx')
```

```python
df7['날짜(전처리)'] = pd.to_datetime(df7['날짜'])
```

```python
df7['공급월'] = df7['날짜(전처리)'].dt.month
```

```python
pd.pivot_table(data=df7, index='공급월',values='재고량',aggfunc='sum')
```

```python
pd.pivot_table(data=df7, index='공급월',values='재고량',aggfunc='sum').to_excel('result1.xlsx') ## 피봇도 저장이 됨
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', family='Malgun Gothic')

sns.barplot(data=df7, x='공급월', y='재고량', ci =None, estimator =sum)
plt.savefig('img_result.png')
```
